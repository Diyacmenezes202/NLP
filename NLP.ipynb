```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet

# Download necessary resources (if not already done)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to get part of speech for each word
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# Function to perform lemmatization
def lemmatize_sentence(sentence):
    words = word_tokenize(sentence)  # Tokenize the sentence into words
    pos_tags = nltk.pos_tag(words)  # Get part of speech tags for words
    lemmatized_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in pos_tags
    ]  # Lemmatize each word with its POS
    return lemmatized_words

# Interactive loop for user input
while True:
    user_input = input("Enter a sentence (or 'exit' to quit): ")
    if user_input.lower() == 'exit':
        print("Exiting the program.")
        break
    
    lemmatized_sentence = lemmatize_sentence(user_input)

    print("Original Sentence:", user_input)
    print("Lemmatized Words:", lemmatized_sentence)
```

```python
# Install necessary libraries
!pip install nltk matplotlib seaborn

# Import required libraries
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
from nltk.tokenize import word_tokenize
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

nltk.download('punkt')

# Function for stemming
def stem_sentence(sentence, stemmer):
    """Stem each word in a given sentence using the provided stemmer."""
    words = word_tokenize(sentence)
    stemmed_words = [stemmer.stem(word.lower()) for word in words]
    return stemmed_words

# Initialize the stemmers
port_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()
snowball_stemmer = SnowballStemmer("english")

# Function to evaluate stemming
def evaluate_stemming(user_sentence, expected_output):
    porter_result = stem_sentence(user_sentence, port_stemmer)
    lancaster_result = stem_sentence(user_sentence, lancaster_stemmer)
    snowball_result = stem_sentence(user_sentence, snowball_stemmer)
    
    # Calculate accuracy
    accuracies = {
        "Porter": accuracy_score(expected_output, porter_result),
        "Lancaster": accuracy_score(expected_output, lancaster_result),
        "Snowball": accuracy_score(expected_output, snowball_result),
    }
    
    # Display results
    print("\nEvaluation Results:")
    print("Original Sentence:", user_sentence)
    print("Expected Stems:  ", expected_output)
    print("Porter Result:   ", porter_result, f" (Accuracy: {accuracies['Porter'] * 100:.2f}%)")
    print("Lancaster Result:", lancaster_result, f" (Accuracy: {accuracies['Lancaster'] * 100:.2f}%)")
    print("Snowball Result: ", snowball_result, f" (Accuracy: {accuracies['Snowball'] * 100:.2f}%)")
    
    # Confusion Matrix Visualization
    def plot_confusion_matrix(expected, result, algorithm_name):
        cm = confusion_matrix(expected, result, labels=sorted(set(expected + result)))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=sorted(set(expected + result)), 
                    yticklabels=sorted(set(expected + result)))
        plt.title(f"Confusion Matrix - {algorithm_name}")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.show()
    
    # Plot confusion matrices
    plot_confusion_matrix(expected_output, porter_result, "Porter Stemmer")
    plot_confusion_matrix(expected_output, lancaster_result, "Lancaster Stemmer")
    plot_confusion_matrix(expected_output, snowball_result, "Snowball Stemmer")

# Main loop for user interaction
while True:
    print("\nProvide a sentence for stemming evaluation (type 'exit' to quit)")
    user_sentence = input("Enter a sentence: ")
    if user_sentence.lower() == 'exit':
        print("Exiting the application.")
        break

    # Get expected output from the user
    expected_output = input("Enter the expected stemmed output as a comma-separated list (e.g., cat,run,fast): ").split(',')
    expected_output = [word.strip() for word in expected_output]  # Clean up spaces
    
    # Evaluate the stemming results
    evaluate_stemming(user_sentence, expected_output)
```
